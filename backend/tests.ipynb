```python
keep_going = True
chat_history = []
while keep_going:
    user_input = input('You: ')
    if user_input == 'exit':
        keep_going = False
    else:
        print(user_input)
        chat_history.append(('human',user_input))
        response = chatbot.invoke({'chat_history':chat_history})
        print(response['chat_history'][-1].content)
        chat_history.append(('ai',response['chat_history'][-1].content))
```
